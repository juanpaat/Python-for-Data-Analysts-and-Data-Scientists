# Data Visualisation

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline    
```

```python
titanic = pd.read_csv('Data/train.csv')
```

```python
titanic.head()
```

# Line chart

```python
plt.figure(figsize=(10,3))
plt.plot(titanic['Age'])
plt.title('Age of the Titanic Passengerld')
plt.xlabel('id')
plt.ylabel('Age')
plt.show()
```

# Histogram

```python
plt.figure(figsize=(10,4))
sns.histplot(titanic['Age'])     # displot  or histplot
plt.title('Titanic Histogram of the Age')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(titanic.Age, shade = True)
plt.title('Titanic Histogram of the Age')
plt.show()
```

```python
g = sns.FacetGrid(titanic, col="Survived", height=3)
g.map(sns.kdeplot, 'Age', shade=True)
g.despine(left=True,bottom=True);
```

```python
g = sns.FacetGrid(titanic, col='Survived', row='Pclass', hue='Sex', height=3)
g.map(sns.kdeplot, 'Age', shade=True).add_legend()
g.despine(left=True, bottom=True)
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.boxplot(titanic['Age'])#,orient='v')
plt.title('Titanic Age Boxplot')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Survived',y='Age',data=titanic)
plt.title('Titanic Boxplot of survivors and age')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x='Sex',y='Age',hue='Survived',data=titanic)
plt.title('Titanic Boxplot of survivors, gender and age')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot('Survived',data=titanic)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(y='Age',x='Pclass',data=titanic)
plt.show()
```

```python
import plotly.express as px
fig = px.scatter(titanic,x='Age',y='Fare', color='Sex',size='Pclass')
fig.show()
```
